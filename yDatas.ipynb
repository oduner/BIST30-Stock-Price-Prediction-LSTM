```python
import os
import pandas as pd
import yfinance as yf
from glob import glob
from datetime import datetime, timedelta
```

```python
csv_path = "Bist/*.csv"
csv_files = glob(csv_path)
bist = []

for file in csv_files:
    df = pd.read_csv(file, skiprows=2)
    df["Date"] = pd.to_datetime(df["Date"])
    df.columns = ["Date","Close", "High", "Low", "Open", "Volume"]
    with open(file, 'r') as f:
        second_line = f.readlines()[1]
        code = second_line.split(',')[1].strip()
    
    df.reset_index(drop=True, inplace=True)
    df[["Close", "High", "Low", "Open", "Volume"]] = df[["Close", "High", "Low", "Open", "Volume"]].astype(float)
    df["Change"] = df["Close"].pct_change() * 100
    df["Change"] = df["Change"].astype(float)
    df["OpenChange"] = df["Open"].pct_change() * 100 
    df["OpenChange"] = df["OpenChange"].astype(float)
    df = df.dropna()
    df.reset_index(drop=True, inplace=True)
    df.index.name = code
    
    bist.append(df)
    bist_file_path = os.path.join("Bist", f"{code[:-3]}.csv")
    df.to_csv(bist_file_path, index=False)
    print(f"{code[:-3]} işlenip kaydedildi.")
```

```python
csv_path = "Test/*.csv"
csv_files = glob(csv_path)
test = []

for file in csv_files:
    df = pd.read_csv(file, skiprows=2)
    df["Date"] = pd.to_datetime(df["Date"])
    df.columns = ["Date","Close", "High", "Low", "Open", "Volume"]
    with open(file, 'r') as f:
        second_line = f.readlines()[1]
        code = second_line.split(',')[1].strip()
    
    df.reset_index(drop=True, inplace=True)
    df[["Close", "High", "Low", "Open", "Volume"]] = df[["Close", "High", "Low", "Open", "Volume"]].astype(float)
    df["Change"] = df["Close"].pct_change() * 100
    df["Change"] = df["Change"].astype(float)
    df["OpenChange"] = df["Open"].pct_change() * 100 
    df["OpenChange"] = df["OpenChange"].astype(float)
    df = df.dropna()
    df.reset_index(drop=True, inplace=True)
    df.index.name = code
    
    test.append(df)
    test_file_path = os.path.join("Test", f"{code[:-3]}T.csv")
    df.to_csv(test_file_path, index=False)
    print(f"{code[:-3]}T işlenip kaydedildi.")
```

```python
# Klasörler
bist_path = "Bist/*.csv"
test_path = "Test/*.csv"

# Saat kontrolü
now = datetime.now()
if now.hour < 18:
    print("Saat 18:00'den önce, işlem yapılmayacak.")
    exit()

# Tüm test dosyalarını işleyelim
test_files = glob(test_path)

for test_file in test_files:
    code = os.path.basename(test_file).replace(".csv", "")
    bist_file = test_file.replace("Test", "Bist")

    # Test ve Bist verilerini oku
    test_df = pd.read_csv(test_file)
    bist_df = pd.read_csv(bist_file)

    # Sütun adlarını yazdırarak kontrol et
    print(f"{test_file} sütun adları: {test_df.columns}")
    print(f"{bist_file} sütun adları: {bist_df.columns}")

    # Eğer "Date" sütunu yoksa, ilk sütunu datetime olarak alalım
    if 'Date' not in test_df.columns:
        test_df.columns = ['Date', 'Close', 'High', 'Low', 'Open', 'Volume']
    
    if 'Date' not in bist_df.columns:
        bist_df.columns = ['Date', 'Close', 'High', 'Low', 'Open', 'Volume']

    test_df["Date"] = pd.to_datetime(test_df["Date"])
    bist_df["Date"] = pd.to_datetime(bist_df["Date"])

    last_date = test_df["Date"].max()
    next_date = last_date + timedelta(days=1)

    # Yarının verisi var mı kontrol et
    new_data = yf.download(f"{code}.IS", start=next_date.strftime("%Y-%m-%d"), progress=False)

    if new_data.empty:
        print(f"{code}: Yeni veri bulunamadı.")
        continue

    new_data = new_data.reset_index()[["Date", "Close", "High", "Low", "Open", "Volume"]]
    new_data["Change"] = new_data["Close"].pct_change() * 100
    new_data["OpenChange"] = new_data["Open"].pct_change() * 100
    new_data = new_data.dropna()

    if new_data.empty:
        print(f"{code}: İşlenebilir yeni veri yok.")
        continue

    print(f"{code}: {len(new_data)} satır yeni veri eklenecek.")

    # Veri kaydırma işlemi
    n = len(new_data)

    # Test verisine ekle
    test_df = pd.concat([test_df, new_data], ignore_index=True)

    # Testin başından n satır al, bist'in sonuna ekle
    moved_data = test_df.iloc[:n]
    bist_df = pd.concat([bist_df, moved_data], ignore_index=True)

    # Bist'in başından n satır sil
    bist_df = bist_df.iloc[n:].reset_index(drop=True)

    # Test'in başından n satır sil
    test_df = test_df.iloc[n:].reset_index(drop=True)

    # Kaydet
    test_df.to_csv(test_file, index=False)
    bist_df.to_csv(bist_file, index=False)

print("Güncelleme tamamlandı.")
```
