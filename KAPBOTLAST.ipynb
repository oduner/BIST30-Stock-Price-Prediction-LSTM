```python
import undetected_chromedriver as uc
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.action_chains import ActionChains
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from dotenv import load_dotenv
import os
import glob
import json
```

```python
def initializer(i):
    company_codes = [
        "AKBNK", "AEFES", "ASELS", "ASTOR", "BIMAS", "CIMSA", "EKGYO", "ENKAI", "EREGL", "FROTO",
        "SAHOL", "HEKTS", "KRDMD", "KCHOL", "KOZAL", "MGROS", "PGSUS", "PETKM", "SASA", "TAVHL",
        "TOASO", "TCELL", "TUPRS", "THYAO", "GARAN", "ISCTR", "SISE", "TTKOM", "ULKER", "YKBNK"
    ]
    save_dir = fr"C:\Users\boss\Desktop\Codes\Final Project v.3.0\KAP\{company_codes[i]}"
    os.makedirs(save_dir, exist_ok=True)
    prefs = {
        "savefile.default_directory": save_dir,
        "download.prompt_for_download": False,
        "download.directory_upgrade": True,
        "printing.print_preview_sticky_settings.appState": json.dumps({
            "recentDestinations": [{"id": "Save as PDF", "origin": "local", "account": ""}],
            "selectedDestinationId": "Save as PDF",
            "version": 2
        })
    }

    options = uc.ChromeOptions()
    options.add_experimental_option("prefs", prefs)
    options.add_argument("--kiosk-printing")
    driver = uc.Chrome(options=options)
    driver.get("https://www.kap.org.tr/tr")
    time.sleep(10)
    i = kapbot(company_codes,driver,save_dir,i)
    return i
```

```python
def kapbot(company_codes,driver,save_dir,i):
    for i in range(len(company_codes)):        
        try:
            search_bar = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.ID, "all-search")))
            ActionChains(driver).move_to_element(search_bar).click().perform()
            print("Clicked the 'search_bar'.")
            time.sleep(3)
            search_bar.clear()
            print("Search bar cleared.")
            time.sleep(3)
            search_bar.send_keys(company_codes[i])
            print(f"Typed '{company_codes[i]}' into the search bar.")
            time.sleep(3)
        except Exception as e:
            print("Error clicking the 'search_bar':", e)
            break
        
        try:
            search_bar_btt = WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.XPATH, "//button[contains(@class, 'flex items-center lg:h-auto h-9 justify-center lg:px-31 px-3 lg:py-3 py-3 border-l border-kap-gray bg-kap-gray rounded-[9px] cursor-pointer')]")))
            search_bar_btt.click() 
            print("Clicked the search_bar_btt.")
            time.sleep(3)
        except Exception as e:
            print("Error clicking the search_bar_btt:", e)
            break
        
        try:
            company_first_btt = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, f"//span[contains(text(), '{company_codes[i]}')]")))
            company_first_btt.click() 
            print("Clicked the company_first_btt.")
            time.sleep(3)
        except Exception as e:
            print("Error clicking the company_first_btt:", e)
            break
        
        try:
            notification_btt = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.ID, "notifications-tab")))
            notification_btt.click()
            print("Clicked on the notification_btt.")
            time.sleep(3)
        except Exception as e:
            print("Error clicking on the notification_btt:", e)
            break
        
        try:
            date_chooser = WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.XPATH, "//div[contains(@class, 'MuiSelect-select MuiSelect-outlined MuiInputBase-input MuiOutlinedInput-input css-117jitv')]")))
            date_chooser.click() 
            print("Clicked the date_chooser.")
            time.sleep(3)
        except Exception as e:
            print("Error clicking the date_chooser:", e)
            break
        
        try:
            date_chooser_month = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "//li[contains(@class, 'MuiMenuItem-root') and .//p[text()='Son 1 ay']]")))
            date_chooser_month.click()
            print("Clicked the date_chooser_month.")
            time.sleep(3)
        except Exception as e:
            print("Error clicking the date_chooser_month:", e)
            break
        
        try:
            notification_chooser = WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.XPATH, "//div[contains(@class, 'flex flex-col w-full h-14 lg:w-[500px]')]")))
            notification_chooser.click() 
            print("Clicked the notification_chooser.")
            time.sleep(3)
        except Exception as e:
            print("Error clicking the notification_chooser:", e)
            break
        
        try:
            notification_chooser_all = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "//li[contains(@class, 'MuiMenuItem-root') and .//p[text()='Tüm Bildirimler']]")))
            notification_chooser_all.click()
            print("Clicked the notification_chooser_all.")
            time.sleep(3)
        except Exception as e:
            print("Error clicking the notification_chooser_all:", e)
            break
        
        try:
            apply_choosed = WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.XPATH, "//*[contains(@class, 'filter-btn ml-2 xl:px-4 px-3 py-3 bg-dark rounded-lg lg:block hidden cursor-pointer')]")))
            apply_choosed.click() 
            print("Clicked the apply_choosed.")
            time.sleep(3)
        except Exception as e:
            print("Error clicking the apply_choosed:", e)
            break

        for j in range(1,6):
            try:
                first_notification = WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.XPATH, f"//tr[@id='notification{j}']")))
                time.sleep(1)
                ActionChains(driver).move_to_element(first_notification).click().perform()
                print("Bildirim tıklandı.")
                time.sleep(3)
            except Exception as e:
                print("Bildirim tıklanamadı:", e)
                break
            
            try:
                WebDriverWait(driver, 10).until(EC.number_of_windows_to_be(2))
                new_window = driver.window_handles[-1]
                driver.switch_to.window(new_window)
                print("Yeni sekmeye geçildi.")
                time.sleep(3)
            except Exception as e:
                print("Yeni sekmeye geçilemedi:", e)
                break
            
            try:
                send_date_element = WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.XPATH,"//div[./div[text()='Gönderim Tarihi']]/div/span[@class='mr-3']")))
                send_date_raw = send_date_element.text.strip()
                send_date = send_date_raw.replace(".", "-")
                print("Gönderim Tarihi:", send_date)
                driver.close()
                driver.switch_to.window(driver.window_handles[0])
                print("İlk sekmeye geri dönüldü.")
                time.sleep(3)
            except Exception as e:
                print("Gönderim tarihi alınamadı:", e)
                break

            '''if j != 1:
                try:
                    choosed_clear = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, f"//tr[@id='notification{j-1}']//input[@type='checkbox']")))
                    choosed_clear.click()
                    print("Clicked on the choosed_clear.")
                    time.sleep(5)
                except Exception as e:
                    print("Error clicking on the choosed_clear:", e)
                    break'''
            
            try:
                choosed_notification = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, f"//tr[@id='notification{j}']//input[@type='checkbox']")))
                choosed_notification.click()
                print("Clicked on the choosed_notification.")
                time.sleep(3)
            except Exception as e:
                print("Error clicking on the choosed_notification:", e)
                break

            try:
                print_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "//button[contains(@class, 'font-variant-caps-all-small font-extrabold relative block text-text-color lg:text-base text-sm before:absolute before:bg-')]")))
                print_button.click()
                print("Clicked on the print_button.")
                time.sleep(3)
            except Exception as e:
                print("Error clicking on the print_button:", e)
                break

            try:
                clear_all = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "//*[text()='Tümünü Temizle']")))
                clear_all.click()
                print("Clicked on the clear_all.")
                time.sleep(5)
            except Exception as e:
                print("Error clicking on the clear_all:", e)
                break

            
            pdf_file_name = f"{company_codes[i]}_{send_date}_{j}.pdf"
            new_pdf_path = os.path.join(save_dir, pdf_file_name)
            list_of_unrenamed_pdfs = glob.glob(os.path.join(save_dir, "[0-9]*.pdf"))
            if list_of_unrenamed_pdfs:
                latest_unrenamed_pdf = max(list_of_unrenamed_pdfs, key=os.path.getctime)
                if os.path.exists(new_pdf_path):
                    timestamp = int(time.time())
                    base_name, ext = os.path.splitext(pdf_file_name)
                    new_pdf_path = os.path.join(save_dir, f"{base_name}_{timestamp}{ext}")
                
                os.rename(latest_unrenamed_pdf, new_pdf_path)
                print(f"PDF başarıyla yeniden adlandırıldı: {os.path.basename(new_pdf_path)}")
            else:
                print("Henüz adlandırılmamış PDF dosyası bulunamadı.")
    return i
```
